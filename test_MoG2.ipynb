```python
import cv2
from torchvision import transforms as T

vidcap = cv2.VideoCapture("/home/mahsa/gmm_dae/dataset/archive3/Normal/Normal_Videos_003_x264.mp4")
bs_MoG = cv2.createBackgroundSubtractorMOG2(history=30)

transform_x = T.Compose([T.ToTensor(), T.Normalize(mean=0.5, std=0.5)])

while vidcap.isOpened():
    ret, frame = vidcap.read()

    # Check if the frame is valid
    if not ret or frame is None or frame.shape[0] == 0 or frame.shape[1] == 0:
        break

    
    mask = bs_MoG.apply(frame)
    # mask = transform_x(mask)
    # mask = mask.numpy()
    # mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)


    cv2.imshow('video', frame)
    cv2.imshow('mask', mask)

    # Exit the loop when 'q' is pressed
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
vidcap.release()
```

```python
while vidcap.isOpened():
    ret, frame = vidcap.read()

    # Check if the frame is valid
    if not ret or frame is None or frame.shape[0] == 0 or frame.shape[1] == 0:
        break
    
    mask = bs_MoG.apply(frame)
    mask = transform_x(mask)
    mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)

    cv2.imshow('video', frame)

    # Exit the loop when 'q' is pressed
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

print("shape of the frames:", frame.shape)
cv2.imshow('mask', mask)  # Display the last mask outside the loop
cv2.destroyAllWindows()
vidcap.release()
```
